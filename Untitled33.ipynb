###Given the dataset's numerous columns (features), we intend to perform feature selection based on the correlation coefficient.

```python
test_data.corr()
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    matrix=dataset.corr()
    for i in range(len(matrix.columns)):
        for j in range(i):
            if abs(matrix.iloc[i, j]) > threshold:
                colname = matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_fea=correlation(test_data.corr(),0.8)
```

```python
len(corr_fea)
```

```python
test_data.drop(corr_fea,axis=1,inplace=True)
```

###To visualize anomaly regions in the time series data, we will initially convert the multivariate time series data into univariate using PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

####Before applying PCA (Principal Component Analysis), we will standardize our data.

```python
scaler = StandardScaler()
scaled_time_series_data = scaler.fit_transform(test_data)
```

```python
pca = PCA(n_components=1)
```

```python
pca.fit(scaled_time_series_data)
```

```python
transformed_data = pca.transform(scaled_time_series_data)
```

```python
transformed_data
```

###After applying PCA and obtaining the transformed data, we will convert the resultant data back to a DataFrame format

```python
transformed_df = pd.DataFrame(transformed_data, columns=['feature'])
```

```python
transformed_df.head()
```

```python
transformed_df.tail()
```

```python
transformed_df.plot(figsize=(10,10))
```

###We will concatenate the test data and test labels.

```python
final_df=pd.concat([transformed_df,test_label],axis=1)
```

```python
final_df.rename(columns={'0': 'is_anomaly'}, inplace=True)
```

```python
final_df.head()
```

```python
final_df['is_anomaly'].value_counts()
```

###We will separate the data into anomaly and non-anomaly categories for the final plot

```python
inlier_df = final_df.loc[final_df['is_anomaly'] == 0]
anomaly_df = final_df.loc[final_df['is_anomaly'] == 1]
```

```python
plt.figure(figsize=(30,30))
fig, ax = plt.subplots()

ax.scatter(inlier_df.index, inlier_df['feature'], color='blue', s=3, label='Inlier')
ax.scatter(anomaly_df.index, anomaly_df['feature'], color='red', label='Anomaly')
ax.legend(loc=2)

plt.grid(False)
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

##Let's determine the variables that serve as the root cause for the anomalies observed in our data.





###We will employ supervised machine learning techniques to identify the variables that serve as the root cause for the anomalies in our dataset

```python
X=test_data
```

```python
X.head()
```

```python
y=final_df['is_anomaly']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred1=rfc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

###After observing the nature of the data, we have decided to proceed with the Random Forest Classifier for further analysis and anomaly detection.

```python
feature_importance = rfc.feature_importances_
```

###We will visualize the feature importance of the selected Random Forest Classifier to understand which variables are most influential in detecting anomalies

```python
plt.figure(figsize=(10, 6))
plt.barh(X.columns, feature_importance, color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance Analysis')
plt.show()
```
