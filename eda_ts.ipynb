###TIME RESAMPLING

```python
df.resample(rule='A').min()
```

this is giving us minimum values for every specific years

```python
df.resample(rule='A').max()
```

also its a data frame

```python
#lets plot
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
## LETS CHECKOUT QUATERLY
```

```python
df.resample(rule='QS').max()
```

```python
df.resample(rule='QS').max()['High'].plot()
```

```python
df.resample(rule='A').mean()['Open'].plot(kind='bar')
```

IF YOU WANT MONTHLY JUST CHANGE THE RULE TO 'M'

```python
df.head()
```

```python
df['Open'].rolling(window=10,min_periods=1).mean()
```

min_periods ensures that you get rolling means starting from the first observation itself, rather than waiting for the window size number of observations to be available

```python
df['Open:30_days_rolling']=df['Open'].rolling(window=30,min_periods=1).mean()
```

```python
df
```

```python
df[['Open','Open:30_days_rolling']].plot(figsize=(12,5))
```

```python
#Simple Moving Average (it is basically used for smoothening of the graph)
```

```python
df['Open'].plot(figsize=(15,6))
```

```python
#lots of zigzag movement
```

```python
df['Open:50_days_rolling']=df['Open'].rolling(window=50,min_periods=1).mean()
```

```python
df['Open:10_days_rolling']=df['Open'].rolling(window=10,min_periods=1).mean()
df['Open:20_days_rolling']=df['Open'].rolling(window=20,min_periods=1).mean()
```

```python
df
```

```python
df[['Open','Open:10_days_rolling','Open:20_days_rolling','Open:30_days_rolling','Open:50_days_rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(12,5))
```

###CUMULATIVE MOVING AVERAGE


```python
df['CMA']=df['Close'].expanding().mean()
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
```

###EXPONENTIAL MOVING AVERAGE/ EXPONENTIAL WEIGHTED MOVING AVERAGE

```python
df['EMA']=df['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df[['CMA','EMA']].plot(figsize=(15,5))
```
