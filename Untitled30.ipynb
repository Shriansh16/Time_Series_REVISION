###Given the dataset's numerous columns (features), we intend to perform feature selection based on the correlation coefficient.

```python
test_data.corr()
```

```python
def correlation(dataset,threshold):
    col_corr=set()   # so that repeated value naa aa paye
    matrix=dataset.corr()
    for i in range(len(matrix.columns)):
        for j in range(i):
            if abs(matrix.iloc[i, j]) > threshold:
                colname = matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_fea=correlation(test_data.corr(),0.8)
```

```python
len(corr_fea)
```

```python
test_data.drop(corr_fea,axis=1,inplace=True)
```

###To visualize anomaly regions in the time series data, we will initially convert the multivariate time series data into univariate using PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

#####Before applying PCA (Principal Component Analysis), we will standardize our data.

```python
scaler = StandardScaler()
scaled_time_series_data = scaler.fit_transform(test_data)
```

```python
pca = PCA(n_components=1)
```

```python
pca.fit(scaled_time_series_data)
```

```python
transformed_data = pca.transform(scaled_time_series_data)
```

```python
transformed_data
```

####After applying PCA and obtaining the transformed data, we will convert the resultant data back to a DataFrame format

```python
transformed_df = pd.DataFrame(transformed_data, columns=['feature'])
```

###We will separate the data into anomaly and non-anomaly categories for the final plot

#####We will apply the Augmented Dickey-Fuller test to determine whether our time series data is stationary or not

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adf_test(final_df['feature'])
```

####Hence, our time series data is stationary, indicating a constant mean.
