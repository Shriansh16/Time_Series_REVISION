###Given the dataset's numerous columns (features), we intend to perform feature selection based on the correlation coefficient.

```python
test_data.corr()
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    matrix=dataset.corr()
    for i in range(len(matrix.columns)):
        for j in range(i):
            if abs(matrix.iloc[i, j]) > threshold:
                colname = matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_fea=correlation(test_data.corr(),0.8)
```

```python
len(corr_fea)
```

```python
test_data.drop(corr_fea,axis=1,inplace=True)
```

#####We will apply the Augmented Dickey-Fuller test to determine whether our time series data is stationary or not

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adf_test(final_df['feature'])
```

####Hence, our time series data is stationary, indicating a constant mean.

##Let's determine the variables that serve as the root cause for the anomalies observed in our data.

####We will employ supervised machine learning techniques to identify the variables that serve as the root cause for the anomalies in our dataset

```python
X=test_data
```

```python
X.head()
```

```python
y=final_df['is_anomaly']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

###We will assess the performance of various models to determine their effectiveness in identifying the variables that contribute to anomalies.

```python
models={'rfc':RandomForestClassifier(),'gbc':GradientBoostingClassifier(),'adc':AdaBoostClassifier(),'lr':LogisticRegression(),'svc':SVC()}
```

```python
report={}
for i in range(len(models)):
  model=list(models.values())[i]
  model.fit(X_train,y_train)
  pred1=model.predict(X_test)
  model_name=list(models.keys())[i]
  report[model_name]=accuracy_score(y_test,pred1)
```

```python
report
```

###After evaluating the accuracy scores of multiple models, it is evident that the GradientBoosting Classifier achieves the highest accuracy score of 94%. Therefore, we will select the GradientBoosting Classifier for further analysis and anomaly detection

```python
gbc=GradientBoostingClassifier()
```

```python
gbc.fit(X_train,y_train)
```

```python
y_pred2=gbc.predict(X_test)
```

```python
feature_importance = gbc.feature_importances_
```

###We will visualize the feature importance of the selected GradientBoosting Classifier to understand which variables are most influential in detecting anomalies

```python
plt.figure(figsize=(10, 6))
plt.barh(X.columns, feature_importance, color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance Analysis')
plt.show()
```
